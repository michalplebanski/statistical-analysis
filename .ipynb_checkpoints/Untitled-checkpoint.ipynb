```python
# manipulacja danymi
import numpy as np
import pandas as pd

# wizualizacja danych
import matplotlib.pyplot as plt
import seaborn as sns

# statystyczna analiza danych
from scipy import stats

# przygotowanie danych
from sklearn.preprocessing import StandardScaler, MinMaxScaler
dataset = pd.read_csv('titanic.csv')
dataset
```

```python
fare = dataset["Fare"]
#Średnia
mean_fare = fare.mean()
print(f"Średnia opłata: {mean_fare}")
```

```python
# Obliczenie kwartyli
q1 = fare.quantile(0.25)
q2 = fare.quantile(0.5)
q3 = fare.quantile(0.75)
print(f"1. kwartyl: {q1}, Mediana: {q2}, 3. kwartyl: {q3}")
```

```python
# Rysowanie wykresu pudełkowego
plt.boxplot(fare)
plt.title('Wykres pudełkowy dla opłat')
plt.xlabel('Fare')
plt.ylabel('Wartość')
plt.show()
```
